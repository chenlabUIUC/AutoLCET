Reconstruction of simulated etching trajectories. See "alignment and reconstruction example" for environment requirements.

```python
import pandas
import numpy as np
import tifffile as tif
import svmbir
import re
from pathlib import Path
import os
import matplotlib.pyplot as plt
```

```python
folder = 'simulated_outputs'
data_folders = [p for p in Path(folder).iterdir() if p.is_dir()]
names = [p.name for p in data_folders]
angle_ranges = {'PdAu': (-55, 60+5, 5),
               'PdRu': (-50, 55+2.5, 2.5),
               'Cu3A': (-60, 50+2.5, 2.5),
               'PdPt': (-60, 50+2.5, 2.5),
               'Chir': (-60, 55+2.5, 2.5)}
```

```python
names
```

```python
for i in range(len(names)):
    if names[i]=="Old sims":
        continue

    # if names[i][-3:] != '90s':
    #     continue
        
    tifs = list(data_folders[i].rglob("*/**/*.tif"))
    print(data_folders[i])
    angle_matrix = np.arange(*angle_ranges[names[i][:4]])
    angles = angle_matrix/180*np.pi
    out_dir = 'recon/'+names[i]+'/'
    os.makedirs(out_dir, exist_ok=True)
    for j in range(len(tifs)):
        filename = 'stage_' + str(j+1)
        print("Reconstructing "+filename)
        data = tif.imread(str(data_folders[i])+'/sims/'+filename+'.tif')
        recon = svmbir.recon(data, angles, stop_threshold=0.002, max_iterations=2000,verbose=0)
        tif.imwrite(out_dir+filename+'_recon.tif', recon)
```

```python
from skimage.filters import threshold_isodata, gaussian

errors = {}
sims_dir = 'simulated_outputs/'

for sample in names:
    print(sample)
    
    recon_dir = Path('recon/' + sample)
    recons = sorted(recon_dir.glob("stage_*_recon.tif"), key=lambda p: int(re.search(r"stage_(\d+)_recon", p.name).group(1)))
    gt_idx = np.loadtxt('recon/Idx_' + sample + '.csv', delimiter=",")[:, -1]

    Vs_recon = []
    Vs_gt = []
    errors_V = []
    errors_pw = []
    for i in range(len(recons)):
        print("Analyzing "+str(recons[i]))
        recon = tif.imread(recons[i])
        thresh = threshold_isodata(recon)
        binarized = recon > thresh
    
        binarized = gaussian(binarized, sigma=3, preserve_range=True, truncate=1)
        binarized = binarized>0.5
        
        sims = list(Path(sims_dir+sample).glob("*.tif"))
        gt = tif.imread(sims[int(gt_idx[i]-1)])>0

        V_recon = np.sum(binarized)
        V_gt = np.sum(gt)

        errors_V.append(np.abs(V_recon-V_gt)/V_gt)

        Vs_recon.append(V_recon)
        Vs_gt.append(V_gt)
        
        sL = np.array(binarized.shape)
        sS = np.array(gt.shape)
    
        # compute start indices (centered)
        start = ((sL - sS) / 2).astype(int)
        end = start + sS
    
        # perform slicing
        binarized_cropped = binarized[
            start[0]:end[0],
            start[1]:end[1],
            start[2]:end[2]
        ]

        errors_pw.append(np.sum(np.logical_xor(binarized_cropped,gt))/np.sum(gt))

    errors[sample] = {"V error": np.array(errors_V), "Pixelwise error": np.array(errors_pw), 
                      "V_recon": np.array(Vs_recon), "V_gt":np.array(Vs_gt)}
```
