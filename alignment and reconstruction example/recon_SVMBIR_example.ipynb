# 0. Import 

```python
import pandas
import numpy as np
import tifffile as tif
import svmbir
```

# 1. Load angles

```python
##this block finds the tilt angles
tilt_angle = pandas.read_csv('input/tiltAnglesSeries__20240612_124708.csv',header = None).to_numpy().squeeze()[:,1]
##change the tilt_range to match the number of tilts per tilt series of _20240612_124708_log_norm
tilt_range = 41
angle_matrix = []
lut = []
loc = 0
while ((loc + tilt_range - 1) < tilt_angle.shape[0]):
    angles = tilt_angle[loc:loc + tilt_range]
    angle_matrix.append(angles)
    lut.append(np.array((loc, loc+tilt_range)))
    loc += tilt_range - 1
angle_matrix = np.array(angle_matrix)
lut = np.array(lut)
print(angle_matrix)
print(lut)
```

# 2. SVMBIR reconstruction

```python
##this block does reconstruction
##please change two folder names
for i in list(range(len(angle_matrix))):
    data = tif.imread('output/align/_20240612_124708/' + "{:03d}".format(i) + '.tif')
    angles = angle_matrix[i]/180*np.pi
    recon = svmbir.recon(data, angles, stop_threshold=0.002, max_iterations=2000,verbose=2)
    tif.imwrite('output/reconstruction/_20240612_124708/' + "{:03d}".format(i) + '.tif', recon)
```
